```python
# Importar libreras
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider
import warnings
warnings.filterwarnings("ignore")

# Configuracin de grficos
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Parmetros de la seal
A = 2     # Amplitud (puedes cambiarla)
T = 4     # Perodo en segundos (puedes cambiarlo)
Fs = 200  # Frecuencia de muestreo

# Vector de tiempo para un perodo
tv = np.arange(-T/2, T/2, 1/Fs)
Nm = len(tv)

# Crear la seal triangular original (versin par)
xo = A - (2*A/T) * np.abs(tv)

# Mximo nmero de armnicos a calcular
N = 50
wo = 2 * np.pi / T

# Calcular los coeficientes de Fourier (an y bn)
an = np.zeros(N+1)
bn = np.zeros(N+1)

# a0
an[0] = A / 2
# bn es cero por simetra par

for n in range(1, N + 1):
    if n % 2 != 0:  # n es impar
        an[n] = (4 * A) / ((n**2) * (np.pi**2))
    else: # n es par
        an[n] = 0

# Calcular la potencia de la seal original para el error
Px = A**2 / 3

# Funcin interactiva para la reconstruccin
def plot_reconstruction(Na=1):
    # Reconstruir la seal con Na armnicos
    xe = np.full_like(tv, an[0]) # Empezar con el nivel DC
    for n in range(1, Na + 1):
        xe += an[n] * np.cos(n * wo * tv)
        # No sumamos bn porque son cero

    # Calcular error relativo
    P_harmonics = an[0]**2 + 0.5 * np.sum(an[1:Na+1]**2 + bn[1:Na+1]**2)
    er = 1 - (P_harmonics / Px)

    # Graficar
    plt.figure(figsize=(12, 6))
    plt.plot(tv, xo, color='r', linewidth=4, label='x(t) Original')
    plt.plot(tv, xe, color='b', linewidth=2, label='$\hat{x}(t)$ Reconstruida')
    plt.title(f'Reconstruccin con {Na} armnicos | Error relativo = {100*er:.2f}%', fontsize=16)
    plt.xlabel('t [s]')
    plt.ylabel('x(t)')
    plt.grid(True)
    plt.legend()
    plt.ylim(-0.2, A + 0.2)
    plt.show()

    # Graficar espectro
    plt.figure(figsize=(12, 4))
    n_vector = np.arange(0, N+1)
    plt.stem(n_vector, an, 'b', markerfmt='bo', basefmt=" ")
    plt.title('Espectro de Amplitudes (Coeficientes $a_n$)', fontsize=16)
    plt.xlabel('Armnico (n)')
    plt.ylabel('Amplitud')
    plt.xlim(-0.5, Na + 0.5)
    plt.grid(True)
    plt.show()


interact(plot_reconstruction, Na=IntSlider(min=1, max=N, step=1, value=1));
```
