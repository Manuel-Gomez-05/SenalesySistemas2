Los ritmos cerebrales (también llamados ondas cerebrales u oscilaciones neurales) son patrones repetitivos de actividad eléctrica en el cerebro.

# Visualizacion de espectogramas

consultar que es STFT

La Short-Time Fourier Transform (STFT), o Transformada de Fourier de Tiempo Corto, es una técnica matemática usada para analizar señales que cambian con el tiempo (señales no estacionarias), como el audio, la música o, en tu caso, las señales de EEG.

Para entenderla fácil, piensa en la diferencia entre una foto y un video:

1. El Problema de la Transformada de Fourier (FT) normal
La Transformada de Fourier clásica toma toda tu señal y te dice qué frecuencias existen en total, pero no te dice cuándo ocurrieron.

Analogía: Es como recibir la lista de ingredientes de una canción (tiene notas Do, Re, Mi) pero sin la partitura; no sabes en qué orden se tocan.

2. La Solución: STFT
La STFT divide la señal larga en pequeños pedazos (llamados ventanas) y aplica la Transformada de Fourier a cada pedacito por separado.

Analogía: En lugar de analizar la canción entera de golpe, la cortas en fragmentos de 1 segundo y analizas las notas de cada segundo. Luego pegas esos análisis en orden.

Resultado: Obtienes un mapa de Tiempo vs. Frecuencia (generalmente visualizado como un Espectrograma), donde puedes ver cosas como: "al segundo 5 empezó una onda Alpha".

3. ¿Cómo funciona el "Windowing" (Ventaneo)?
No se corta la señal con tijeras (eso crearía errores en los bordes). Se multiplica la señal por una "función ventana" (como la ventana de Hann o Hamming) que suaviza los bordes a cero.

Solapamiento (Overlap): Las ventanas suelen superponerse (ej. 50%) para no perder información en los bordes que se suavizaron.

4. El compromiso (Trade-off) de Resolución
Aquí hay una regla de oro importante (Principio de Incertidumbre de Heisenberg-Gabor):

Ventana Ancha (larga en tiempo): Tienes excelente resolución de frecuencia (ves muy bien los Hz exactos), pero mala resolución temporal (no sabes exactamente cuándo cambió la señal).

Ventana Estrecha (corta en tiempo): Sabes exactamente cuándo ocurre un evento (buena resolución temporal), pero es borroso saber qué frecuencia exacta es (mala resolución de frecuencia).

#Common Spatial Patterns

Consulté qué son los Common Spatial Patterns (CSP) y su aplicación al procesado de señales EEG

**Los Common Spatial Patterns** (CSP) o Patrones Espaciales Comunes son, probablemente, el algoritmo más famoso y utilizado en el mundo de las Interfaces Cerebro-Computadora (BCI).

Si en tu trabajo anterior con la STFT estabas analizando "cuándo" y "en qué frecuencia" ocurre algo, con CSP analizas "en qué lugar" (espacio) del cerebro ocurre, pero de una forma matemática muy inteligente.

Aquí tienes los detalles:

1. **¿Qué es y para qué sirve?**
Es un algoritmo de filtrado espacial. Su objetivo es tomar la señal de todos tus electrodos (multicanal) y mezclarlos matemáticamente para crear nuevos "canales virtuales" que maximicen la diferencia entre dos tareas mentales.

La analogía: Imagina que tienes 20 micrófonos en un estadio (tus electrodos) y quieres escuchar al entrenador gritando (Clase A) y distinguirlo del ruido del público (Clase B).

CSP crea una combinación de esos 20 micrófonos para que, en la nueva señal resultante, la voz del entrenador se escuche fuertísimamente y el público se silencie, y viceversa.

2. **El Principio Matemático (Varianza)**
El núcleo de CSP es la varianza (que en señales biológicas equivale a la energía/potencia). El algoritmo busca una matriz de proyección tal que:

Para la Clase 1 (ej. imaginar mover la mano derecha), la varianza de la señal filtrada sea máxima.

Para la Clase 2 (ej. imaginar mover la mano izquierda), la varianza de esa misma señal sea mínima.

De esta forma, al clasificador (la IA) le queda muy fácil distinguir: si la energía es alta, es mano derecha; si es baja, es mano izquierda.

3. **Aplicación Estrella: Imaginería Motora (Motor Imagery)**
CSP es el estándar de oro para detectar cuando una persona imagina mover sus extremidades.

Cuando imaginas mover la mano derecha, se produce una desincronización (baja energía en banda Mu/Alpha) en el hemisferio izquierdo del cerebro.

CSP detecta este patrón espacial cruzado (lateralización) mejor que cualquier humano mirando las gráficas.

# Punto 2.2

1. Fundamentos: Filtros FIR e IIR desde la Transformada ZLa función de transferencia de un filtro digital se define como:$$H(z) = \frac{Y(z)}{X(z)} = \frac{\sum_{k=0}^{M} b_k z^{-k}}{1 + \sum_{k=1}^{N} a_k z^{-k}}$$
Filtros FIR (Finite Impulse Response):
Son filtros cuya respuesta al impulso es finita. En la ecuación anterior, los coeficientes $a_k$ son todos cero.

Ecuación Z: $H(z) = \sum_{k=0}^{M} b_k z^{-k}$

**Polos y Ceros:** Solo tienen ceros (los polos están en el origen $z=0$). Esto garantiza que siempre son estables.

Característica clave: Pueden diseñarse para tener fase lineal. Esto significa que retrasan todas las frecuencias por la misma cantidad de tiempo, preservando la forma de la onda.

**Filtros IIR (Infinite Impulse Response):** Son filtros con retroalimentación (feedback). Usan tanto coeficientes $b_k$ (entrada) como $a_k$ (salida anterior).

**Ecuación Z:** Tienen numerador y denominador.

**Polos y Ceros:** Tienen polos que determinan la estabilidad. Si un polo cae fuera del círculo unitario en el plano Z ($|z| > 1$), el filtro se vuelve inestable.

Característica clave: Son muy eficientes. Logran cortes de frecuencia abruptos con un orden de filtro bajo (menos cómputo), pero introducen distorsión de fase (no lineal).

2. **Uso en Patrones de Imaginación Motora (MI):**

En tareas de MI, el cerebro genera patrones específicos conocidos como ERD/ERS (Desincronización/Sincronización Relacionada con Eventos) en las bandas Mu (8-12 Hz) y Beta (13-30 Hz). Así es como los filtros ayudan a describirlos:

A. Patrones Espectrales (Frecuencia)El objetivo es aislar la energía en las bandas Mu y Beta donde ocurre la actividad motora.

Uso de IIR: Debido a que las aplicaciones BCI suelen ser en tiempo real, se prefieren filtros IIR (como Butterworth o Chebyshev) porque requieren menos coeficientes y cómputo para separar, por ejemplo, la banda de 8-12 Hz.

Uso de FIR: Se usan cuando el análisis es offline o se requiere una precisión espectral extrema sin alterar la fase, permitiendo ver la estructura fina del espectro de potencia.

B. **Patrones Temporales (Tiempo):**

Aquí buscamos cuándo ocurre la imaginación del movimiento. La preservación de la forma de onda es vital.

El problema de la Fase: Un filtro IIR tiene fase no lineal. Esto significa que la componente de 8 Hz puede sufrir un retardo diferente a la de 12 Hz, deformando el pico del potencial EEG o cambiándolo de lugar en el tiempo.

La solución FIR: Para estudios donde la latencia exacta del comando cerebral es crítica (ej. medir tiempos de reacción neuronal), se usan filtros FIR de fase lineal. Aunque requieren más orden computacional, aseguran que el patrón temporal (el momento exacto del ERD) no se desplace artificialmente.

C. Patrones Espaciales (Localización)La imaginación motora es contralateral: imaginar mover la mano derecha activa el hemisferio izquierdo (electrodo C3) y viceversa (electrodo C4).

Pre-procesamiento: Aunque los filtros FIR/IIR son temporales, son el paso previo obligatorio para los filtros espaciales.

El Filtro CSP (Common Spatial Patterns): Es el algoritmo estándar para MI. El CSP funciona maximizando la varianza de una clase (mano izq) mientras minimiza la otra.

Interacción: Para que el CSP encuentre el patrón espacial correcto, la señal debe haber sido filtrada previamente (con FIR o IIR) estrictamente en la banda de frecuencia de interés (8-30 Hz). Si filtras mal temporalmente (dejando pasar ruido de 50Hz o artefactos musculares), el filtro espacial no podrá separar las zonas C3 y C4 correctamente.

**Introducción y Marco Teórico:** El Desafío de la VariabilidadEn este proyecto, abordamos la detección de patrones de Imaginación Motora (MI) superando las limitaciones de los métodos tradicionales de banda fija. La principal dificultad en BCI (Brain-Computer Interface) es la alta variabilidad inter-sujeto e intra-sujeto.

1.1 **Influencia de Artefactos Fisiológicos:**

Las señales EEG puras de actividad motora (ritmos $\mu$ y $\beta$) suelen estar contaminadas por fuentes biológicas no cerebrales:

**Artefactos Oculares (EOG):** El parpadeo y los movimientos sacádicos generan potenciales eléctricos mucho mayores que el EEG (>100 $\mu V$).

**Impacto:** Contaminan principalmente las bajas frecuencias ($<4$ Hz), pero sus armónicos pueden afectar la banda $\mu$ (8-12 Hz).Solución: El filtrado paso-banda estricto en nuestro algoritmo elimina la componente DC y bajas frecuencias del EOG.

**Actividad Muscular (EMG):** La tensión en la mandíbula, cuello o frente introduce ruido de alta frecuencia.

Impacto: Se solapa con el ritmo $\beta$ (>15 Hz), confundiendo al clasificador.

Solución: El algoritmo CSP (Common Spatial Patterns) ayuda a aislar estas fuentes si su ubicación espacial (ej. electrodos temporales externos) difiere de la corteza motora central (C3, C4).

1.2 **Factores Genéticos, Cognitivos y de Experiencia**

La "inconsistencia" en la identificación de patrones no es solo ruido, sino
variabilidad intrínseca:

**Variabilidad Anatómica y Genética:** La conductividad del cráneo varía entre individuos, afectando la amplitud de la señal.

La orientación de los dipolos neuronales depende de la morfología de los surcos cerebrales. Un electrodo C3 puede captar la señal máxima en un sujeto, mientras que en otro la señal se proyecta mejor hacia FC3 debido a la geometría de su giro precentral. CSP resuelve esto aprendiendo filtros espaciales personalizados.

2.**Factores Cognitivos y "BCI Illiteracy":**

Se estima que un 15-30% de los usuarios no logran controlar una BCI (analfabetismo BCI).

Experiencia Previa: Usuarios entrenados muestran una desincronización (ERD) más fuerte y localizada.Estrategia Mental: Imaginar "apretar una pelota" vs. "levantar una pesa" activa redes ligeramente distintas.
