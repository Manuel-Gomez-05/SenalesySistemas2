Cuaderno señales estandar SIMPY


```python
#Use la función  rect(t)  para construir una función  cos(ωt)  con  ω=2 , la cual se enciende en  t=−12  y se apaga en  t=+12 .
# definir la funcion
omega=2
t=sym.symbols('t')
rectified_cosine = sym.cos(omega * t) * rect(t)


#graficar
sym.plot(rectified_cosine.subs(omega, 2), (t, -2, 2), ylim=[-1.2, 1.2], ylabel=r'$\cos(\omega t) \cdot \text{rect}(t)$')
```

```python
#Genere un rectificador de onda completa utilizando la función  sgn(t)
# Definir
sine_wave = sym.sin(omega * t)

# señal
rectified_sine = sine_wave * sym.sign(sine_wave)

#darle valor a omega y dibujar con plot
sym.plot(sine_wave.subs(omega, 2), (t, -2*sym.pi, 2*sym.pi), ylabel=r'Original Signal: $\sin(\omega t)$')

# dibujar señal rectificada
sym.plot(rectified_sine.subs(omega, 2), (t, -2*sym.pi, 2*sym.pi), ylabel=r'Rectified Signal: $| \sin(\omega t) |$', line_color='red')
```
